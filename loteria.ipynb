```python
from random import randint
from pprint import pprint
```

```python
def loteria(quantidade_de_jogos, tipojogo=1):
    """
        megasena = 1
        lotofacil = 2
    """
    if tipojogo == 1:
        maiornumero = 60
        quantidadenumeros = 6

    elif tipojogo == 2:
        maiornumero = 25
        quantidadenumeros = 15

    lista_total = []
    for jogos in range(quantidade_de_jogos):
        lista = []

        while True:

            x = randint(1,maiornumero)

            if x not in lista:
                lista.append(x)

            if len(lista) == quantidadenumeros:
                break

        lista_total.append(sorted(lista))
    return lista_total

pprint(loteria(500,2))

```
